```python
people = {
    "Name": ["Gautam", "Kumaresan", "Devi"],
    "Age": [26, 61, 53],
    "Gender": ["Male", "Male", "Female"]
}
import pandas as pd
df = pd.DataFrame(people)
df
```

```python
df['Gender'] == 'Male'
```

```python
filt = (df['Gender'] == 'Male')
df[filt]
```

```python
# Since we can do column filtering as well while using loc so it is preferred
df.loc[filt, 'Name']
```

```python
# And (&)/ OR (|) operators
filt1 = (df['Gender'] == 'Male') & (df['Name'] == 'Gautam')
df.loc[filt1]
```

```python
filt2 = (df['Gender'] == 'Male') | (df['Name'] == 'Gautam')
df.loc[filt2]
```

```python
#not condition
df.loc[~filt2]
```
